```python
df_shortage_nursing_b = df_covid['Shortage of Nursing Staff'] == 'Y'
df_shortage_nursing = df_covid[df_shortage_nursing_b]
df_shortage_nursing_bins = df_shortage_nursing.groupby(['Provider Name'])['Week Ending'].count().reset_index(name='Number of Nursing Shortage weeks')
df_shortage_nursing_bins = df_shortage_nursing_bins.rename(columns = {"Provider Name" : "Facility"})
df_shortage_nursing_bins['Number of Nursing Shortage weeks'] = pd.cut(x=df_shortage_nursing_bins['Number of Nursing Shortage weeks'], bins=[0,10,20,float("inf")], labels=['0 - 10', '11 - 20', '>20'])
df_shortage_nursing_bins
```

```python
df_covid_occupancy = df_covid[(df_covid['Number of All Beds'] != 'NaN')]
df_covid_occupancy_f = df_covid_occupancy[(df_covid_occupancy['Total Number of Occupied Beds'] != 'NaN')]
df_covid_occupancy_f['Percentage Occupancy'] = df_covid_occupancy_f['Total Number of Occupied Beds'] * 100 /df_covid_occupancy_f['Number of All Beds']
df_covid_occupancy_middlesex = df_covid_occupancy_f[(df_covid_occupancy_f['Provider State'] == 'MA') & (df_covid_occupancy_f['County'] == 'Middlesex')]
df_covid_occupancy_middlesex[['Provider Name','Week Ending','Percentage Occupancy']]
```

```python
# 100% or more occupancy --- More than 100% shows that the nursing home facility was overtaxed and special provisions were made for COVID-19
df_covid_occupancy_middlesex_morethan100 = df_covid_occupancy_middlesex[(df_covid_occupancy_middlesex['Percentage Occupancy'] >= 100)]
df_covid_occupancy_middlesex_morethan100['Provider Name'].unique()
```

```python
# 100% or More Occupancy for 3 Weeks or more ----
df_covid_occupancy_middlesex_morethan100_3weeks = df_covid_occupancy_middlesex_morethan100[df_covid_occupancy_middlesex_morethan100['Provider Name'].map(df_covid_occupancy_middlesex_morethan100['Provider Name'].value_counts()) >= 3]
df_covid_occupancy_middlesex_morethan100_3weeks['Provider Name'].value_counts()
```
